```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
#Draw a line in diagram from position (0,10) to postion (5,150):
xpoints= np.array([1,2])
ypoints= np.array([5,10])


plt.plot(xpoints,ypoints)
plt.show()
```

```python
xpoints= np.array([1,6])
ypoints= np.array([4,5])


plt.plot(xpoints,ypoints,"*")
plt.show()
```

```python

ypoints= np.array([3,8,1,12])
xpoints=np.array([5,6,7,9])
plt.plot(xpoints,ypoints,marker="*" , ms=10,color="green")
plt.show()
```

```python
#without taking x points
ypoints=np.array([3,8,1,10])
plt.plot(ypoints,marker='o', ms=10,color='red')
plt.show()
```

```python
#edge color
ypoints=np.array([2,8,2])
plt.plot(ypoints,marker="*", ms=10, mec='y',color="pink")
plt.show()
```

```python
#markerfacecolor
ypoints=np.array([7,6,3,8])
plt.plot(ypoints,marker="*", ms="11" , mfc='g' ,mec='k')
plt.show()
```

```python
#linestyle-dashed,linewidth
ypoints=np.array([6,3,9,27,8])
plt.plot(ypoints,marker="o",linestyle='dashed',mec='hotpink',ms=10, mfc='blue', linewidth=3)
plt.show()
```

```python
#linewidth or lw
#when color is not given the default color is blue
ypoints=np.array([7,6,3,8])
plt.plot(ypoints,lw=3)
plt.show()
```

```python
#two different plots
y1=np.array([6,1,3,9,0])
y2=np.array([2,1,8,6,9])
plt.plot(y1,color='red')
plt.plot(y2,"k")
plt.show()
```

```python
#labels #title
x=np.array([2,4,6,10,5,33,66,88,99])
y=np.array([49,90,77,7,9,4,23,87,12])
plt.plot(x, y , color="y",linewidth=2 )

plt.xlabel('Average Pulse', fontsize=20, color='red')
plt.ylabel('Calorie Burnage', fontsize=20, color='b')
plt.title("Sample",fontsize=20,color='b')
plt.show()

```

```python
#grid
x=np.array([1000,3000,4000,7000,20000,40000,67000,123000,])
y=np.array([2002,2004,2008,2010,2015,2020,2022,2023])
plt.plot(x, y , color="y",linewidth=2 )

plt.xlabel('Viewers', fontsize=20, color='red')
plt.ylabel('year', fontsize=20, color='red')
plt.title("Anime Trend in India",fontsize=20,color='b')
plt.grid(color='g', linestyle='dotted')
plt.show()
```

```python
#scatter
x=np.array([1,2,3,4,5,6])
y=np.array([5,10,20,25,30,35])
plt.scatter(x,y)
plt.show()
```

```python
#Bar
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.bar(x,y,color='blue')
plt.show()
```

```python
x=np.array(["A","B","C","D"])
y=np.array([3,8,1,10])
plt.barh(x,y,color='blue')
plt.show()
```

```python
x=np.array([10,20,30,40,70])

plt.hist(x,color='blue')
plt.show()
```

```python
#piechart
y=np.array([10,30,135,34,56])
myexplode=[0,0.4,0.5,0.8,0.7]
mycolors=['green','black','pink','red','blue']
plt.pie(y,labels=mylabels ,startangle=30 ,explode=myexplode, colors=mycolors)
mylabels=['A','B','C','D','E']

plt.show()
```

```python
#legend
y=np.array([35,45,69,76])
mylabels=['A','B','C','D']
plt.pie(y)
plt.legend(labels=mylabels)
plt.show()
```
